```python
import pandas as pd
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
```

```python
df = pd.read_csv('/content/wine_dataset.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df['style'] = df['style'].replace('red', 0)
```

```python
df['style'] = df['style'].replace('white', 1)
```

```python
#separando as variáveis em preditoras e variável alvo

y = df['style']
x = df.drop('style', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#criando conjunto de dados de treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
```

```python
x_treino.shape
#y_treino.shape
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
#criação do modelo

modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

#imprimindo resultados

resultado = modelo.score(x_teste, y_teste)
print('Acurácia:', resultado)
```
